# Mendeteksi  Kontur

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# gambar asli

```python
image = cv2.imread('build.jpeg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
edged = cv2.Canny(image_rgb, 30, 200)
contours, _ = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
image_copy = image.copy()
```

# gambar contour

```python
image_copy = image.copy()
cv2.drawContours(image_copy, contours, -1, (0, 0, 255), 3)
```

# gambar blank

```python
binary = np.zeros_like(image_rgb)
cv2.drawContours(binary, contours, -1, (255, 255, 255), thickness=cv2.FILLED)
```

```python
blank_image = np.zeros((1500,1500,3), np.uint8)
blank_image[:,0:1500//2] = (255,255,255)
blank_image[:,1500//2:1500] = (255,255,255)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original Image")

ax[1].imshow(cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB))
ax[1].set_title("Drawn Contours")

ax[2].imshow(binary, cmap='binary')
ax[2].set_title("Blank Image")


plt.tight_layout()
plt.show()
```
